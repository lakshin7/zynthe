```python
import pandas as pd
import json
import os

# Make sure the 'data' folder exists
os.makedirs("data", exist_ok=True)
def csv_to_jsonl(csv_path, jsonl_path):
    df = pd.read_csv(csv_path)
    # For quick testing, you can take a subset (e.g., 2k samples) if needed
    df = df.sample(n=2000, random_state=42)
    with open(jsonl_path, 'w') as f:
        for _, row in df.iterrows():
            f.write(json.dumps({
                "text": row['review'],
                "label": int(row['sentiment'] == 'positive')
            }) + "\n")

csv_to_jsonl("IMDB Dataset.csv", "data/imdb_train.jsonl")  # or train split
csv_to_jsonl("IMDB Dataset.csv", "data/imdb_val.jsonl")    # or val split
```
